```python
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
```

```python
sd.fit_transform(df)
```

```python
X = df.drop('target', axis = 1)
X.head()
```

```python
y = df['target']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , train_size= 0.7, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score ,confusion_matrix, classification_report
```

```python
accuracy_lr = accuracy_score(y_test, y_pred)
accuracy_lr
```

```python
precision_lr = precision_score(y_test, y_pred)
precision_lr
```

```python
recall_lr = recall_score(y_test, y_pred)
recall_lr 
```

```python
f1_score_lr = f1_score(y_test, y_pred)
f1_score_lr
```

```python
confusion_matrix_lr = confusion_matrix(y_test, y_pred)
confusion_matrix_lr
```

```python
classification_report_lr = classification_report(y_test, y_pred)
classification_report_lr
```

```python
report = classification_report(y_test, y_pred, output_dict=True)
df_report_lr = pd.DataFrame(report).transpose()
print(df_report_lr)
```

```python
accuracy_dt = accuracy_score(y_test, y_pred)
accuracy_dt
```

```python
precision_dt = precision_score(y_test, y_pred)
precision_dt
```

```python
recall_dt = recall_score(y_test, y_pred)
recall_dt 
```

```python
f1_score_dt = f1_score(y_test, y_pred)
f1_score_dt
```

```python
confusion_matrix_dt = confusion_matrix(y_test, y_pred)
confusion_matrix_dt
```

```python
classification_report_dt = classification_report(y_test, y_pred)
classification_report_dt
```

```python
report = classification_report(y_test, y_pred, output_dict=True)
df_report_dt = pd.DataFrame(report).transpose()
print(df_report_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```
